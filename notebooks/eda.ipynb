# Zomato Restaurant Dataset - Exploratory Data Analysis

This notebook explores the Zomato restaurant dataset to understand:
- Dataset structure and characteristics
- Distribution of restaurants across locations
- Rating and cost patterns
- Cuisine popularity
- Budget tier analysis

```python
# Import required libraries
import sys
sys.path.append('..')

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from src.data_loader import ZomatoDataLoader

# Set display options
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', 100)

# Set plotting style
plt.style.use('seaborn-v0_8-darkgrid')
sns.set_palette('husl')

print("✓ Libraries imported successfully")
```

## 1. Load the Dataset

```python
# Initialize data loader
loader = ZomatoDataLoader(config_path='../config.yaml')

# Load and process data
df = loader.load_and_process()

print(f"Dataset loaded: {len(df)} restaurants")
print(f"Columns: {list(df.columns)}")
```

## 2. Dataset Overview

```python
# Display basic information
print("Dataset Shape:", df.shape)
print("\nData Types:")
print(df.dtypes)
print("\nMemory Usage:")
print(df.memory_usage(deep=True))
```

```python
# Display first few rows
df.head(10)
```

```python
# Check for missing values
missing = df.isnull().sum()
missing_pct = (missing / len(df)) * 100

missing_df = pd.DataFrame({
    'Missing Count': missing,
    'Percentage': missing_pct
})

print("Missing Values Analysis:")
print(missing_df[missing_df['Missing Count'] > 0])
```

```python
# Summary statistics
df.describe()
```

## 3. Location Analysis

```python
# Number of restaurants per location
location_counts = df['location'].value_counts()

print(f"Total unique locations: {len(location_counts)}")
print("\nTop 15 locations by restaurant count:")
print(location_counts.head(15))
```

```python
# Visualize top locations
plt.figure(figsize=(12, 6))
location_counts.head(15).plot(kind='bar')
plt.title('Top 15 Locations by Number of Restaurants', fontsize=14, fontweight='bold')
plt.xlabel('Location', fontsize=12)
plt.ylabel('Number of Restaurants', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

## 4. Rating Analysis

```python
# Rating statistics
print("Rating Statistics:")
print(f"Mean Rating: {df['aggregate_rating'].mean():.2f}")
print(f"Median Rating: {df['aggregate_rating'].median():.2f}")
print(f"Std Dev: {df['aggregate_rating'].std():.2f}")
print(f"Min: {df['aggregate_rating'].min():.1f}, Max: {df['aggregate_rating'].max():.1f}")
```

```python
# Distribution of ratings
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Histogram
axes[0].hist(df['aggregate_rating'], bins=30, edgecolor='black', alpha=0.7)
axes[0].axvline(df['aggregate_rating'].mean(), color='red', linestyle='--', 
                label=f'Mean: {df["aggregate_rating"].mean():.2f}')
axes[0].axvline(df['aggregate_rating'].median(), color='green', linestyle='--', 
                label=f'Median: {df["aggregate_rating"].median():.2f}')
axes[0].set_title('Distribution of Restaurant Ratings', fontsize=14, fontweight='bold')
axes[0].set_xlabel('Rating', fontsize=12)
axes[0].set_ylabel('Frequency', fontsize=12)
axes[0].legend()

# Box plot
axes[1].boxplot(df['aggregate_rating'], vert=True)
axes[1].set_title('Rating Box Plot', fontsize=14, fontweight='bold')
axes[1].set_ylabel('Rating', fontsize=12)

plt.tight_layout()
plt.show()
```

```python
# Rating categories
rating_bins = [0, 2.5, 3.5, 4.0, 5.0]
rating_labels = ['Poor (0-2.5)', 'Average (2.5-3.5)', 'Good (3.5-4.0)', 'Excellent (4.0+)']
df['rating_category'] = pd.cut(df['aggregate_rating'], bins=rating_bins, labels=rating_labels)

rating_dist = df['rating_category'].value_counts()
print("\nRating Categories:")
print(rating_dist)

# Pie chart
plt.figure(figsize=(8, 8))
plt.pie(rating_dist, labels=rating_dist.index, autopct='%1.1f%%', startangle=90)
plt.title('Restaurant Rating Distribution by Category', fontsize=14, fontweight='bold')
plt.show()
```

## 5. Cost Analysis

```python
# Cost statistics
print("Cost Statistics (Average for Two):")
print(f"Mean Cost: ₹{df['average_cost_for_two'].mean():.2f}")
print(f"Median Cost: ₹{df['average_cost_for_two'].median():.2f}")
print(f"Std Dev: ₹{df['average_cost_for_two'].std():.2f}")
print(f"Min: ₹{df['average_cost_for_two'].min():.0f}, Max: ₹{df['average_cost_for_two'].max():.0f}")
```

```python
# Distribution of costs
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Histogram
axes[0].hist(df['average_cost_for_two'], bins=50, edgecolor='black', alpha=0.7)
axes[0].axvline(df['average_cost_for_two'].mean(), color='red', linestyle='--', 
                label=f'Mean: ₹{df["average_cost_for_two"].mean():.0f}')
axes[0].set_title('Distribution of Restaurant Costs', fontsize=14, fontweight='bold')
axes[0].set_xlabel('Average Cost for Two (₹)', fontsize=12)
axes[0].set_ylabel('Frequency', fontsize=12)
axes[0].legend()

# Log scale histogram (better for skewed data)
axes[1].hist(df['average_cost_for_two'], bins=50, edgecolor='black', alpha=0.7)
axes[1].set_yscale('log')
axes[1].set_title('Cost Distribution (Log Scale)', fontsize=14, fontweight='bold')
axes[1].set_xlabel('Average Cost for Two (₹)', fontsize=12)
axes[1].set_ylabel('Frequency (log)', fontsize=12)

plt.tight_layout()
plt.show()
```

## 6. Budget Tier Analysis

```python
# Budget distribution
budget_counts = df['budget'].value_counts()
print("Budget Tier Distribution:")
print(budget_counts)
print(f"\nPercentages:")
print((budget_counts / len(df) * 100).round(2))
```

```python
# Visualize budget distribution
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Bar chart
budget_counts.plot(kind='bar', ax=axes[0], color=['green', 'orange', 'red'])
axes[0].set_title('Restaurant Count by Budget Tier', fontsize=14, fontweight='bold')
axes[0].set_xlabel('Budget Tier', fontsize=12)
axes[0].set_ylabel('Number of Restaurants', fontsize=12)
axes[0].set_xticklabels(axes[0].get_xticklabels(), rotation=0)

# Pie chart
axes[1].pie(budget_counts, labels=budget_counts.index, autopct='%1.1f%%', 
            colors=['green', 'orange', 'red'], startangle=90)
axes[1].set_title('Budget Tier Percentage', fontsize=14, fontweight='bold')

plt.tight_layout()
plt.show()
```

```python
# Average rating by budget tier
budget_rating = df.groupby('budget')['aggregate_rating'].agg(['mean', 'median', 'std', 'count'])
print("\nRating Statistics by Budget Tier:")
print(budget_rating)

# Visualize
plt.figure(figsize=(10, 6))
df.boxplot(column='aggregate_rating', by='budget', figsize=(10, 6))
plt.title('Rating Distribution by Budget Tier', fontsize=14, fontweight='bold')
plt.suptitle('')  # Remove default title
plt.xlabel('Budget Tier', fontsize=12)
plt.ylabel('Rating', fontsize=12)
plt.show()
```

## 7. Cuisine Analysis

```python
# Extract all unique cuisines
if 'cuisines' in df.columns:
    all_cuisines = []
    for cuisines_str in df['cuisines'].dropna():
        for cuisine in str(cuisines_str).split(','):
            all_cuisines.append(cuisine.strip())
    
    from collections import Counter
    cuisine_counts = Counter(all_cuisines)
    
    print(f"Total unique cuisines: {len(cuisine_counts)}")
    print("\nTop 20 Most Popular Cuisines:")
    for cuisine, count in cuisine_counts.most_common(20):
        print(f"{cuisine}: {count}")
```

```python
# Visualize top cuisines
if 'cuisines' in df.columns:
    top_cuisines = pd.Series(dict(cuisine_counts.most_common(15)))
    
    plt.figure(figsize=(12, 6))
    top_cuisines.plot(kind='barh')
    plt.title('Top 15 Most Popular Cuisines', fontsize=14, fontweight='bold')
    plt.xlabel('Number of Restaurants', fontsize=12)
    plt.ylabel('Cuisine', fontsize=12)
    plt.tight_layout()
    plt.show()
```

## 8. Correlation Analysis

```python
# Select numeric columns for correlation
numeric_cols = ['aggregate_rating', 'average_cost_for_two']
if 'votes' in df.columns:
    numeric_cols.append('votes')

correlation_matrix = df[numeric_cols].corr()

plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', center=0, 
            square=True, linewidths=1, cbar_kws={"shrink": 0.8})
plt.title('Correlation Matrix', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.show()
```

```python
# Scatter plot: Cost vs Rating
plt.figure(figsize=(10, 6))
plt.scatter(df['average_cost_for_two'], df['aggregate_rating'], alpha=0.3)
plt.title('Cost vs Rating', fontsize=14, fontweight='bold')
plt.xlabel('Average Cost for Two (₹)', fontsize=12)
plt.ylabel('Rating', fontsize=12)
plt.tight_layout()
plt.show()

# Calculate correlation
corr = df['average_cost_for_two'].corr(df['aggregate_rating'])
print(f"\nCorrelation between Cost and Rating: {corr:.3f}")
```

## 9. Top Restaurants

```python
# Top rated restaurants
print("Top 10 Highest Rated Restaurants:")
top_rated = df.nlargest(10, 'aggregate_rating')[['name', 'location', 'cuisines', 
                                                    'aggregate_rating', 'average_cost_for_two', 'budget']]
display(top_rated)
```

```python
# Most expensive restaurants
print("\nTop 10 Most Expensive Restaurants:")
most_expensive = df.nlargest(10, 'average_cost_for_two')[['name', 'location', 'cuisines', 
                                                            'aggregate_rating', 'average_cost_for_two', 'budget']]
display(most_expensive)
```

```python
# Best value restaurants (high rating, low cost)
print("\nBest Value Restaurants (Rating > 4.0, Cost < ₹500):")
best_value = df[(df['aggregate_rating'] > 4.0) & (df['average_cost_for_two'] < 500)]
best_value = best_value.nlargest(10, 'aggregate_rating')[['name', 'location', 'cuisines', 
                                                            'aggregate_rating', 'average_cost_for_two']]
display(best_value)
```
